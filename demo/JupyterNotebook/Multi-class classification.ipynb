```python
import statsmodels.api as st
iris = st.datasets.get_rdataset('iris','datasets')
y = iris.data.Species
x = iris.data.ix[:, 0:4]
x = st.add_constant(x, prepend = False)
```

```python
y.unique()
```

```python
iris.data
```

```python
mdl = st.MNLogit(y, x)
mdl_fit = mdl.fit(method='bfgs')
print (mdl_fit.summary())
```

```python
mdl_fit.predict(x)
```
