### 1 Document Classification

(1) Convert the "description" of each case to a vector (using the pretrained word to vector dictionary "PubMed-and-PMC-w2v.bin"). You may use the average of word vectors as the document vector representation.

```python
from gensim.test.utils import common_texts
from gensim.models import Word2Vec
import gensim.downloader as api
from gensim.models import TfidfModel
from gensim.corpora import Dictionary
import pandas as pd
from gensim.models import KeyedVectors
import numpy as np
import nltk
from sklearn import preprocessing
from sklearn import svm
from sklearn.metrics import confusion_matrix

mt = pd.read_csv("medicaltranscriptions.csv")
mt_token=[]
lenlist=[]
mt["description"]
```

```python
model=KeyedVectors.load_word2vec_format("PubMed-and-PMC-w2v.bin",binary=True)

description_vec=np.empty((0, 200), float)

for i in range(len(mt["description"])):
    word2vec_ls=np.empty((0, 200), float)
    mt_token_pos=nltk.pos_tag(nltk.word_tokenize(mt["description"][i]))
    for word, pos in mt_token_pos:
        if (pos == 'NN' or pos == 'NNP' or pos == 'NNS' or pos == 'NNPS'):
            try:
                word2vec_ls=np.vstack([word2vec_ls, model.word_vec(word)])
            except:
                pass
    if np.any(np.isnan(word2vec_ls)) or len(word2vec_ls)==0:
        desc_vec=np.zeros((1, 200))
    else:
        desc_vec = np.mean(word2vec_ls, axis=0).reshape(1, 200)
    description_vec=np.vstack([description_vec, desc_vec])

print(description_vec)
```

(2) Build a multi-class classification SVMmodel to classify a document into one of the "medical_specialty"  categories. Use the document vector as your predictors and use "medical_specialty" as your target variable.

```python
Doc2VecRep=pd.DataFrame(description_vec)

le = preprocessing.LabelEncoder()
le.fit(mt['medical_specialty'])
y=le.transform(mt['medical_specialty'])
Doc2VecRep["y"] = y
clf = svm.SVC(kernel='linear')
clf.fit(Doc2VecRep.iloc[:, : 200], Doc2VecRep["y"])
y_pred=pd.Series(clf.predict(Doc2VecRep.iloc[:, : 200]))
```

(3) What are the in-sample recall rates for each document types?

```python
C_matrix = confusion_matrix(np.array(y_pred), np.array(Doc2VecRep["y"]))
pd.DataFrame(C_matrix)
```

```python
rates = np.diag(C_matrix)/np.sum(C_matrix,axis=0)
rates
```

(4) Which "medical_ specialty" has the highest recall rate?

```python
highest_medical_specialty=le.inverse_transform([np.argmax(rates)])
print("medical_specialty with the highest recall rate is:",highest_medical_specialty,"\nThe recall rate is:",max(rates))
```
